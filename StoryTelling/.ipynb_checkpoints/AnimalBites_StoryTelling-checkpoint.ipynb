## 1.2 Data Wrangling

```python
# Convert bite_date to datetime
df['bite_date'] = pd.to_datetime(df['bite_date'], errors='coerce')

# Extract year, month, and day of week
df['year'] = df['bite_date'].dt.year
df['month'] = df['bite_date'].dt.month
df['days_of_week'] = df['bite_date'].dt.day_name()

df = df.dropna(subset=['bite_date'])
```

```python
df.head()
```

```python
df.describe()
```

This dataset is supposed to cover the period from 1985 to 2017, but it contains dates that fall outside of this range. While these could be simple typographical errors, the incorrect dates cannot be accurately corrected, so the corresponding rows will be removed.

```python
start_date = '1985-01-01'
end_date = '2017-12-31'

df[(df['bite_date'] < start_date) | (df['bite_date'] > end_date)]
```

```python
df = df[(df['bite_date'] >= start_date) & (df['bite_date'] <= end_date)]
```

# 2. Data Analysis

## 2.1 Trends in Animal Bite Incidents

```python
yearly_bite = df.groupby('year').size()
monthly_bite = df.groupby('month').size()
days_of_week_bite = df.groupby('days_of_week').size()
```

```python
# Plotting Yearly bite incidents
plt.figure(figsize=(6,4))
yearly_bite.plot(kind='line', color = 'orange')
plt.title("Yearly Animal Bite Incidents")
plt.xlabel('Year')
plt.ylabel('Number of Bites')
plt.grid(True)
```

The data is based on reported animal bites, which may not occur at regular intervals. This irregularity means that the dataset does not have the continuous time-based structure typical of time seTries data. In addition, the number of bites reported could be influenced by external factors like public awareness campaigns, changes in reporting protocols, or even seasonal trends, but these do not imply a consistent, predictable pattern over time. Therefore, this dataset is not suitable to find the trends in time.

## 2.2 Species and Breed Risk

```python
df_clean = df.dropna(subset=['SpeciesIDDesc', 'BreedIDDesc'])

# Frequency Analysis
species_counts = df_clean['SpeciesIDDesc'].value_counts()
breed_counts = df_clean['BreedIDDesc'].value_counts()

# proportion: Relative Frequency
species_prop = species_counts/species_counts.sum() * 100
breed_prop = breed_counts/breed_counts.sum() * 100

# Species Bar chart
species_prop.plot(kind='bar', color='orange', title='Bite Incidents by Species')
plt.xlabel('Species')
plt.ylabel('Proportion of Incidents (%)')

for i, count in enumerate(species_prop):
    plt.text(i, count + 1, str(round(count,2)), ha='center', va='bottom', fontsize=6)
    
plt.show()

# Breed Bar Chart
colors = ['orange' if i < 5 else 'black' for i in range(15)]

breed_prop.head(15).plot(kind='bar', color=colors, title='Top 15 Breeds Involved in Bite Incidents')
plt.xlabel('Breed')
plt.ylabel('Proportion of Incidents (%)')

for i, count in enumerate(breed_prop.head(15)):
    plt.text(i, count, str(round(count,2)), ha='center', va='bottom', fontsize=6)
    
plt.show()
```

## 2.5 Effectiveness of Preventive Actions

### 2.5.1 Advice Issued vs. Incident Outcome

```python
# Convert 'AdvIssuedYNDesc' and 'DispositionIDDesc' to categorical
df['AdvIssuedYNDesc'] = df['AdvIssuedYNDesc'].astype('category')
df['DispositionIDDesc'] = df['DispositionIDDesc'].astype('category')

# Filter out rows where either 'AdvIssuedYNDesc' or 'DispositionIDDesc' is NaN
df_advice_outcome = df.dropna(subset=['AdvIssuedYNDesc', 'DispositionIDDesc'])

# Count the occurrences of each disposition outcome based on whether advice was issued
advice_outcome_counts = df_advice_outcome.groupby(['AdvIssuedYNDesc', 'DispositionIDDesc']).size().unstack().fillna(0)

# Plot stacked bar chart
colors = ['orange', 'black', 'gray', 'blue', 'green']  # Adjust colors as needed
ax = advice_outcome_counts.plot(kind='bar', stacked=True, color=colors, edgecolor='black')
plt.xlabel('Advice Issued')
plt.ylabel('Number of Incidents')
plt.title('Outcome of Incidents Based on Advice Issued')

# Display the exact counts on the bars
for i, row in advice_outcome_counts.iterrows():
    for j, count in enumerate(row):
        plt.text(j, count + 1, f'{int(count)}', ha='center', va='bottom', fontsize=10)

plt.legend(title='Disposition Outcome')
plt.show()
```

```python
# Calculate proportions to evaluate the impact of advice
advice_outcome_proportions = advice_outcome_counts.div(advice_outcome_counts.sum(axis=1), axis=0) * 100

# Plot proportions as a stacked bar chart
ax = advice_outcome_proportions.plot(kind='bar', stacked=True, color=colors, edgecolor='black')
plt.xlabel('Advice Issued')
plt.ylabel('Proportion of Incidents (%)')
plt.title('Proportion of Incident Outcomes Based on Advice Issued')

# Display the exact proportions on the bars
for i, row in advice_outcome_proportions.iterrows():
    for j, proportion in enumerate(row):
        plt.text(j, proportion + 1, f'{proportion:.1f}%', ha='center', va='bottom', fontsize=10)

plt.legend(title='Disposition Outcome')
plt.show()
```
