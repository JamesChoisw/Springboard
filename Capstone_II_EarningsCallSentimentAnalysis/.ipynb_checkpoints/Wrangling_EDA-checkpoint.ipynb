```python
import requests
import pandas as pd
from tqdm import tqdm

# Define the base URL and headers for the GitHub API
base_url = "https://api.github.com/repos/Vxtr10/Earnings-Call-NLP-Strategy-V2/contents/sectors"
headers = {'Accept': 'application/vnd.github.v3+json'}

# Define the list of sectors
sectors = [
    "utilities", "transportation", "telecom", "tech", "software", "semiconductors", 
    "retail", "real-estate", "pharmaceuticals-biotech", "media", "materials", 
    "insurance", "household", "healthcare", "food-beverage-tobacco", "energy", 
    "diversified-financials", "consumer-services", "consumer-retailing", 
    "consumer-durables", "commercial-services", "capital-goods", "banks", 
    "automobiles"
]

# List of S&P 500 tickers
spx_list = tickers

# Prepare a list to hold the data
data_list = []

# Function to get file list from GitHub
def get_file_list_from_github(sector, ticker):
    url = f"{base_url}/{sector}/{ticker}"
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        return response.json()  # Return list of files
    elif response.status_code == 404:
        return []  # Skip if folder doesn't exist
    else:
        print(f"Error: {response.status_code} at {url}")
        return []

# Total number of iterations for progress tracking
total_iterations = len(sectors) * len(spx_list)

# Iterate through sectors and tickers with a progress bar
with tqdm(total=total_iterations, desc="Processing sectors and tickers") as pbar:
    for sector in sectors:
        for ticker in spx_list:
            files = get_file_list_from_github(sector, ticker)
            
            # Check if there are files to process
            if files:
                for file in files:
                    if file['name'].endswith('.csv'):
                        download_url = file['download_url']
                        try:
                            # Download the CSV file
                            df = pd.read_csv(download_url, header=None)
                            
                            # Extract the first four rows (as vectors) and reshape
                            extracted_data = df.iloc[:4, :].transpose()  # Transpose to get each row as a column
                            extracted_data.insert(0, 'Ticker', ticker)  # Insert Ticker column

                            # Append to the data list
                            data_list.append(extracted_data)
                        except Exception as e:
                            print(f"Error processing {file['name']}: {e}")
            
            # Update the progress bar after processing each ticker
            pbar.update(1)

# Combine all extracted data into a single DataFrame
if data_list:
    final_df = pd.concat(data_list, ignore_index=True)

    # Save the combined DataFrame to a CSV file
    final_df.to_csv('data/combined_earnings_calls.csv', index=False)
    print("Data has been combined and saved to 'combined_earnings_calls.csv'")
else:
    print("No data to combine.")
```

```python
fake_news[fake_news['text'].str.contains(r'\(Reuters\)', na=False)]
```

__More than 99% of the true_news dataset contains '(Reuters)' in the text column, while over 99% of the fake_news dataset does not. Therefore, it is possible to achieve over 99% accuracy in classification based solely on whether '(Reuters)' is present in the text column. As a result, we will remove the leading '(Reuters) - ' from all rows in the text column of true_news, and for fake_news, we will delete any rows that contain '(Reuters)' in the text column.__

```python
true_news['text'] = true_news['text'].str.replace(r'^.*\(Reuters\)\s*-\s*', '', regex=True)
fake_news = fake_news[~fake_news['text'].str.contains(r'\(Reuters\)', na=False)]
```

__* Counts and Length__

```python
df['text_length'] = df['text'].str.len()
df.head(10)
```

```python
def word_count(text):
    word_list = text.split()
    return len(word_list)
```

```python
df['word_count'] = df['text'].apply(word_count)
df.head()
```

__Mean Word and Sentence Length__

```python
df['mean_word_len'] = df['text'].map(lambda txt: np.mean([len(word) for word in txt.split()]))
df.head()
```

```python
df['mean_sentence_len'] = df['text'].map(lambda txt: np.mean([len(sentence) for sentence in tokenize.sent_tokenize(txt)]))
df.head()
```

```python
plt.figure(figsize=(16,10))

plt.subplot(2, 2, 1)
sns.violinplot(x='target', y = 'text_length', data=df)
plt.title('Distribution of Text Length by Target')
plt.xlabel('Target')
plt.ylabel('Text Leneth')

plt.subplot(2, 2, 2)
sns.violinplot(x='target', y='word_count', data=df)
plt.title('Distribution of the Number of Words by Target')
plt.xlabel('Target')
plt.ylabel('Num of Words')

plt.subplot(2, 2, 3)
sns.violinplot(x='target', y='mean_word_len', data=df)
plt.title('Distribution of Average Word Length by Target')
plt.xlabel('Target')
plt.ylabel('Avg Word Length')

plt.subplot(2, 2, 4)
sns.violinplot(x='target', y='mean_sentence_len', data=df)
plt.title('Distribution of Average Sentence Length by Target')
plt.xlabel('Target')
plt.ylabel('Avg Sentence Length')
```
